# Laboratory setup

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Parameters

```python
# Width and height of individual panels of the ASIST tank
PANEL_WIDTH = 0.78
PANEL_HEIGHT = 1

# We have a total of 19 panels
ASIST_WIDTH = 19 * PANEL_WIDTH
ASIST_HEIGHT = PANEL_HEIGHT

# Laser Elevation Gauges in Donelan et al. (2004)
D04_LEG1 = 4.592
D04_LEG3 = 8.991

# Pressure ports in Donelan et al. (2004)
D04_P1 = 3.014
D04_P2 = 7.012
D04_P3 = 11.009

# Resting water depth [m]
H = 0.42

# UDM positions in Curcic and Haus (2020)
UDM = [0.89, 2.19, 5.31, 6.04, 8.55, 9.93]

# Pressure ports in Curcic and Haus (2020)
P1 = 0.08
P2 = 13 * PANEL_WIDTH + 0.08

# Hot-film, IRGASON, and Nortek Vectrino positions in Curcic and Haus (2020)
hotfilm = (4.66, H + 0.29)
irgason = (6.2, H + 0.29)
vectrino = (7.82, 0.045)
```

```python
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(111, xlim=(-1, ASIST_WIDTH + 1), ylim=(0, ASIST_HEIGHT))

for u in UDM:
    plt.plot(u, 1, 'rs', ms=8, clip_on=False, zorder=5)
    plt.plot([u, u - 0.1], [1, H], 'k:', lw=1)
    plt.plot([u, u + 0.1], [1, H], 'k:', lw=1)
    
plt.plot(hotfilm[0], hotfilm[1], 'rx', ms=10, clip_on=False, zorder=5)
plt.plot(irgason[0], irgason[1], 'r*', ms=10, clip_on=False, zorder=5)
plt.plot(vectrino[0], vectrino[1], 'rv', ms=10, clip_on=False, zorder=5)

# water
plt.plot([-0.5, ASIST_WIDTH], [H, H], 'c-', lw=2)
plt.fill_between([-0.5, ASIST_WIDTH], [H, H], color='c', alpha=0.5)

# beach
plt.plot([13 * PANEL_WIDTH, ASIST_WIDTH], [0, 0.5], 'k:', lw=3)
plt.text(12, -0.4, 'Porous beach', fontsize=16)

# inlet
plt.plot([-1, 0], [0.5, 0.5], 'k-', lw=3)
plt.plot([-0.5, -0.5], [0, 0.5], 'k-', lw=2)
plt.plot([-1, -0.5], [0.25, 0.25], 'k-', lw=2)
plt.arrow(-1.25, 0.75, 0.75, 0, width=0.1, clip_on=False, zorder=5)
plt.text(-1, 1.2, 'Wind inlet', zorder=10, fontsize=16, ha='center')
plt.text(-1, -0.4, 'Wave paddle', fontsize=16, ha='center')

# outlet
plt.plot([ASIST_WIDTH, ASIST_WIDTH], [0, 0.5], 'k-', lw=3)
plt.plot([ASIST_WIDTH, ASIST_WIDTH + 1], [0.5, 0.5], 'k-', lw=3)
plt.arrow(ASIST_WIDTH, 0.75, 0.75, 0, width=0.1, clip_on=False, zorder=5)
plt.text(ASIST_WIDTH, 1.2, 'Outlet', fontsize=16)

# tank
plt.plot([-1, ASIST_WIDTH + 1], [1, 1], 'k-', lw=3, clip_on=False)
plt.plot([-1, ASIST_WIDTH], [0, 0], 'k-', lw=3, clip_on=False)

# high-speed side cameras
plt.fill_between([0.9, 2.9], y1=[0.22, 0.22], y2=[0.62, 0.62], color='r', alpha=0.3, zorder=50)
plt.fill_between([4.5, 6.5], y1=[0.22, 0.22], y2=[0.62, 0.62], color='r', alpha=0.3, zorder=50)
plt.fill_between([8.1, 10.1], y1=[0.22, 0.22], y2=[0.62, 0.62], color='r', alpha=0.3, zorder=50)

plt.plot([0.6, 0.9], [-0.4, 0.22], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.plot([0.6, 0.9], [-0.4, 0.62], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.plot([0.6, 2.9], [-0.4, 0.22], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.plot([0.6, 2.9], [-0.4, 0.62], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.text(0.8, -0.6, 'Camera-imaging', color='r', fontsize=12, transform=ax.transData)

plt.plot([4.2, 4.5], [-0.4, 0.22], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.plot([4.2, 4.5], [-0.4, 0.62], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.plot([4.2, 6.5], [-0.4, 0.22], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.plot([4.2, 6.5], [-0.4, 0.62], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)

plt.plot([7.8, 8.1], [-0.4, 0.22], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.plot([7.8, 8.1], [-0.4, 0.62], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.plot([7.8, 10.1], [-0.4, 0.22], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)
plt.plot([7.8, 10.1], [-0.4, 0.62], 'r-', zorder=100, lw=1, alpha=0.3, clip_on=False)

# label
plt.text(-3, 2.7, 'Donelan et al. (2004)', fontsize=16)
plt.text(-3, 2.4, 'This study', color='r', fontsize=16)

plt.plot(D04_LEG1, 0, 'k^', ms=16)
plt.plot(D04_LEG3, 0, 'k^', ms=16)

plt.plot([D04_LEG1, D04_LEG1], [0, -0.8], 'k-', clip_on=False)
plt.plot([D04_LEG3, D04_LEG3], [0, -0.8], 'k-', clip_on=False)
plt.plot([D04_LEG1, D04_LEG3], [-0.8, -0.8], 'k-', clip_on=False)
plt.text((D04_LEG1 + D04_LEG3) / 2, -0.85, \
    r'$\dfrac{\partial h}{\partial x}$, $\dfrac{\partial S_{xx}}{\partial x}$',\
    ha='center', va='top', fontsize=16)

plt.plot(D04_P1, 1, 'k.', ms=16, clip_on=False)
plt.plot(D04_P3, 1, 'k.', ms=16, clip_on=False)

plt.plot(P1, 1, 'r.', ms=16, clip_on=False, zorder=5)
plt.plot(P2, 1, 'r.', ms=16, clip_on=False, zorder=5)

plt.plot([D04_P1, D04_P1], [1, 1.2], 'k-', clip_on=False)
plt.plot([D04_P3, D04_P3], [1, 1.2], 'k-', clip_on=False)
plt.plot([D04_P1, D04_P3], [1.2, 1.2], 'k-', clip_on=False)
plt.text(D04_P2, 1.25, r'$\dfrac{\partial p}{\partial x}$', ha='left', va='bottom', fontsize=16)

plt.plot([P1, P1], [1, 1.8], 'r-', clip_on=False)
plt.plot([P2, P2], [1, 1.8], 'r-', clip_on=False)
plt.plot([P1, P2], [1.8, 1.8], 'r-', clip_on=False)
plt.text((P1 + P2) / 2, 1.85, r'$\dfrac{\partial p}{\partial x}$', ha='left', va='bottom', color='r', fontsize=16)

plt.plot([UDM[0], UDM[0]], [1, 2.4], 'r-', clip_on=False)
plt.plot([UDM[4], UDM[4]], [1, 2.4], 'r-', clip_on=False)
plt.plot([UDM[0], UDM[4]], [2.4, 2.4], 'r-', clip_on=False)
plt.text((UDM[0] + UDM[4]) / 2, 2.45, r'$\dfrac{\partial h}{\partial x}$, $\dfrac{\partial S_{xx}}{\partial x}$', ha='left', va='bottom', color='r', fontsize=16)


plt.plot(np.nan, np.nan, 'k^', ms=12, label='Laser Elevation Gauge')
plt.plot(np.nan, np.nan, 'ko', ms=12, label='Air pressure port')
plt.plot(np.nan, np.nan, 'rx', ms=12, label='X-film anemometer')
plt.plot(np.nan, np.nan, 'r*', ms=12, label='Sonic anemometer')
plt.plot(np.nan, np.nan, 'rs', ms=12, label='Ultrasonic Distance Meter')
plt.plot(np.nan, np.nan, 'rv', ms=12, label='Nortek Vectrino')

plt.legend(bbox_to_anchor=(0.9, 0.9), bbox_transform=plt.gcf().transFigure)

plt.xticks(range(16))

plt.xlabel('Fetch [m]', fontsize=16)
plt.ylabel('Height [m]', fontsize=16)

fig.suptitle('Instrument locations and momentum budget components in ASIST', fontsize=16)
fig.subplots_adjust(bottom=0.3, top=0.5)
```
